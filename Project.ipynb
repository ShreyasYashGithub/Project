```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
num_samples = 100
healthy_diet = np.random.normal(loc=10, scale=2, size=num_samples)
unhealthy_diet = np.random.normal(loc=5, scale=3, size=num_samples)
plt.hist([healthy_diet, unhealthy_diet], bins=15, color=['green', 'red'], label=['Healthy Diet', 'Unhealthy Diet'])
plt.title('Dietary Patterns Analysis')
plt.xlabel('Diet Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
studies = ['Study 1', 'Study 2', 'Study 3', 'Study 4', 'Study 5']
effect_sizes = [0.75, 0.82, 0.69, 0.91, 0.78] 
lower_bound = [0.65, 0.75, 0.60, 0.85, 0.70] 
upper_bound = [0.85, 0.89, 0.78, 0.97, 0.86]  
y_pos = np.arange(len(studies))[::-1]
plt.figure(figsize=(8, 6))
plt.errorbar(effect_sizes, y_pos, xerr=[effect_sizes[i] - lower_bound[i] for i in range(len(studies))], fmt='o', color='blue', capsize=5, label='Effect Size')
plt.yticks(y_pos, studies)
plt.axvline(x=1, color='red', linestyle='--')
plt.xlabel('Odds Ratio')
plt.title('Junk Food Consumption and Symptoms of Mental Health Problems: Meta-Analysis')
plt.legend()
plt.grid(axis='x')
plt.show()
```
